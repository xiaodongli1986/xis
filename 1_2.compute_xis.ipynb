### Visulize the xi(s) in six bins

```python
X = range(50);
X = [x+0.5 for x in X]

for omw in [(0.31,-1.0), (0.06,-1.0), (0.41,-1.0), (0.31,-2.5), (0.31,0.0)]:
    om, w = omw
    fig = plt.figure(figsize=(14,7));
    ax1 = fig.add_subplot(121);ax2 = fig.add_subplot(122);

    s1 = 10
    #X = X[s1:len(X)]
    for catname in ['DR12v4-LOWZ', 'DR12v4-CMASS']:
        for ibin in range(3):
            filename = Tpcfrltfilename(cosmoconvertedfilename(binsplittedfilename(datafile(catname), ibin+1), om, w), 51, 51, 120)
            legname = catname + '_bin' +str(ibin+1)
            Y = smu_xis(filename, make_plot=False)
            #Y = Y[s1-1:len(Y)]
            ax1.plot(X, Y, label = legname, lw=2)
            Ynorm = normto1(Y);
            ax2.plot(X, Ynorm, label = legname, lw=2)
    for ax in [ax1,ax2]:
            ax.grid();ax.set_xlabel('s', fontsize=15); ax.set_ylabel('xi(s)', fontsize=15);
            ax.legend(loc='best', frameon=False)
    ax2.set_ylabel('xi(s) normed', fontsize=15);
    ax1.set_title('Cosmology = '+str(om)+' '+str(w))
    fig.savefig('3.xis_data.'+omwstr(om,w)+'.png', format='png')
```
